### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Loading DataSet

```python
dataset = pd.read_csv("Salary_Data.csv")
dataset.info()
```

Splitting into Independatn and dependant Column

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(X)
```

```python
y
```

As There has no missing value we don't have to handle thisüòç

### Splitting Data into Training and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 2)
```

## Now Training the Simple Linear Regression Model on the traning set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# predicting The test result

y_pred = regressor.predict(x_test)
y_pred
```

### Now visualizing the Training set results

```python
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, regressor.predict(x_train), color = "green")
plt.title("Salary vs Experience (On Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Plotting the Test set result

```python
plt.scatter(x_test, y_test, color = "black")
plt.plot(x_train, regressor.predict(x_train), color = "green")
plt.title("Salary vs Experience (On Test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
